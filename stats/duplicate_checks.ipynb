```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import json
import re
from collections import Counter
```

```python
dfs = []
parent_file_location = os.path.dirname(os.path.abspath(os.getcwd()))
latnecy_data_path = os.path.join(parent_file_location,'results','throughput')
for root, dir,files in os.walk(latnecy_data_path):
    for file in files:
        if file == 'raw_data.csv':
            file_path = os.path.join(root, file)
            df = pd.read_csv(file_path)
            dfs.append(df)
```

```python

pub_keys_group = []
for df_to_check in dfs:
    df_to_check.head()
    df_updated = pd.DataFrame()

    df_updated['payload'] = df_to_check['payload'].str.strip('"').apply(lambda x: re.sub(r"'", '"', x))
    df_updated['payload'] = df_updated['payload'].apply(json.loads)
    df_updated = pd.json_normalize(df_updated['payload'])
    df_updated.head()

    pub_keys = []
    for index, row in df_updated.iterrows():
        for batch in row['data']:
            for deposit_data in batch['deposit_data']: 
                pub_keys.append(deposit_data['pubkey'])
                print(deposit_data['pubkey'])
    pub_keys_group.append(pub_keys)


            
```

```python
for pub_keys_itr in pub_keys_group:
    counter = Counter(pub_keys_itr)
    duplicates = [item for item, count in counter.items() if count > 1]
    print(duplicates)
```
